### Map Pointing Different Cities in India
In this project notebook, **'Folium'** package of python is used to do Map pointing. We first create the Map of India. Then point out some important cities of India such as Kolkata, Delhi, Chennai, Mumbai, Bangalore, Indor, Ahmedabad, Hyderabad,Agra.
Using their Latitude and Longitude, the whole project will be completed.

##### The Latitude and Longitude of India and other cities are given below -

* India: **latitude - 20.593683 , longitude - 78.962883**
* Kolkata: **latitude - 22.572645 , longitude - 88.363892**
* Delhi: **latitude - 28.704060 , longitude - 77.102493**
* Chennai: **latitude - 13.082680 , longitude - 80.270721**
* Mumbai: **latitude - 19.075983 , longitude - 72.877655**
* Bangalore: **latitude - 12.971599 , longitude - 77.594566**
* Indor: **latitude - 22.719568 ,  longitude - 75.857727**
* Ahmedabad: **latitude - 23.022505  , longitude - 72.571365**
* Hydrabad: **latitude - 17.385044 , longitude - 78.486671**
* Agra: **latitude - 27.176670 , longitude - 78.008072**

```python
# Importing required packages
import folium
```

```python
# Plotting the map that is India centered
india_map = folium.Map(location = [20.593683,78.962883], zoom_start = 6)
india_map
```

```python
# Adding Multiple Cities on India
locations = [
    {'location': [22.572645,88.363892], 'popup': 'Kolkata'},
    {'location': [28.704060,77.102493], 'popup': 'Delhi'},
    {'location': [13.082680,80.270721], 'popup': 'Chennai'},
    {'location': [19.075983,72.877655], 'popup': 'Mumbai'},
    {'location': [12.971599,77.594566], 'popup': 'Bangalore'},
    {'location': [22.719568,75.857727], 'popup': 'Indor'},
    {'location': [23.022505,72.571365], 'popup': 'Ahmedabad'},
    {'location': [17.385044,78.486671], 'popup': 'Hyderabad'},
    {'location': [27.176670,78.008072], 'popup': 'Agra'}
]

for loc in locations:
  folium.Marker(location = loc['location'], popup = loc['popup']).add_to(india_map)

india_map
```
